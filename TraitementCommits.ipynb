```python
# import required module
from pathlib import Path
import pandas as pd
```

```python
# assign directory
directory = 'D:\PFEMaster\Dataset et pretraitement\Dataset_VCC'
```

```python
 # iterate over files in that directory
files = Path(directory).glob('*.diff')
```

```python
#devide the patches into pre-modification and post-modification versions
def patchDevision (file):
  print (file.name)
  postchange, prechange="","",
  with open (file, "r",encoding="utf-8", errors="ignore") as fr:
    for i, line in enumerate (fr):
        if (len(line.split()) != 0) :
          if line.startswith("@@") ==False:
            if line.startswith("+"):
              subline=line[1:]
              if (len (subline.split()) != 0) :
                postchange+=line
            elif line.startswith("-") :
              subline=line[1:]
              if (len(subline.split()) != 0) :
                prechange+=line
            else:
              postchange+=line
              prechange+=line
    with open("D:/DatasetVCCPlus/"+file.name[:-5]+".txt", "w",encoding="utf-8", errors="ignore") as fw1:
        fw1.write(postchange)
    with open("D:/DatasetVCCMinus/"+file.name[:-5]+".txt", "w",encoding="utf-8", errors="ignore") as fw2:
        fw2.write(prechange)
```

```python
for file in files:
  patchDevision (file)
```
